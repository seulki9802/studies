```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

import folium
```

```python
url = urlopen('https://calop123.github.io/web1.html/img_test2.html')
html = BeautifulSoup(url, 'html.parser')
```

```python
m_n = html.findAll('tr')

infos_dummy = []
for m in m_n:
    n = m.select('td')
    
    infos = []
    for info in n:
        infos += [info.text]
        
    infos_dummy += [infos]
```

```python
from collections import Counter

birth_places = []
death_places = []
for info in infos_dummy:
    birth_places += [info[2] + '/' + info[3]]
    death_places += [info[5] + '/' + info[6]]

birth_frequency = Counter(birth_places).most_common()
death_frequency = Counter(death_places).most_common()
```

```python
import folium

'''지도그리기'''
m = folium.Map(
location = [35.17718451089109, -20.502206518216088]
# , tiles = 'Stamen Terrain'
, zoom_start = 2.5
)


'''마크추가'''
'''빈도 나타내기'''
for birth in birth_frequency:
    
    birth_lat_lon = birth[0].split('/')[1].split(',')
    birth_country = birth[0].split('/')[0]
    
    if birth_lat_lon != ['']:
        folium.CircleMarker(
            birth_lat_lon,
            popup =
            '<p>' + birth_country + '</p>',
            radius = 7 * birth[1],            
            color = '#cc3131',
            fill_color = '#cc3131'
        ).add_to(m)
        
for death in death_frequency:
    
    death_lat_lon = death[0].split('/')[1].split(',')
    death_country = death[0].split('/')[0]
    
    if death_lat_lon != ['']:
        folium.CircleMarker(
            death_lat_lon,
            popup =
            '<p>' + death_country + '</p>',
            radius = 7 * death[1],            
            color = '#000000',
            fill_color = '#000000'
        ).add_to(m)

'''작가 나타내기'''
for info in infos_dummy: 
    #출생지
    if info[3] != '':
        folium.CircleMarker(
            info[3].split(','), # 출생국 위도 경도
            popup =
                '<img src=' + info[7] +' width="100" height="100">' #작가 이미지
                +'<h4> <a href="http://cslkr.iptime.org/Seulki/Author/books/' + info[0].replace(' ','') + '.html">' + info[0] + '</a> </h4>'
                +'<p>'+info[1]+'</p>' # 출생년도
                +'<p>'+info[2]+'</p>', # 출생지
            radius = 7,
            color = '#cc3131',
            fill_color = '#cc3131'
        ).add_to(m)

    #사망한 곳
    if info[6] != '':
        
        folium.CircleMarker(
            info[6].split(','), # 사망한 곳 위도 경도
            popup = 
            '<img src=' + info[7] +' width="100" height="100">' #작가 이미지
            +'<h4> <a href="http://cslkr.iptime.org/Seulki/Author/books/' + info[0].replace(' ','') + '.html">' + info[0] + '</a> </h4>'
            +'<p>'+info[4]+'</p>' # 사망년도
            +'<p>'+info[5]+'</p>', # 사망한 곳
            radius = 7,
            color = '#000000',
            fill_color = '#000000'
        ).add_to(m)


m.save('C:\\Bitnami\\wampstack-7.4.14-0\\apache2\\htdocs\\SeulKi\\2_test\\testmap2.html')


# <h4> <a href=""Edward Abbey </h4>
```
